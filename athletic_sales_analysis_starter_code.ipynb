### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

groupby_athletic_df=athletic_sales_2021_2021.groupby(['region','state','city'])['units_sold'].sum()

groupby_athletic_df.rename({'units_sold':"Total_Products_Sold"})
# Show the top 5 results.
mostProducts_sold=groupby_athletic_df.sort_values(ascending=False)

mostProducts_sold.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.

atheletic_sales_pivot=pd.pivot_table(athletic_sales_2021_2021,values='units_sold',index=['region','state','city'],aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
atheletic_sales_pivot=atheletic_sales_pivot.rename(columns={'units_sold':'Total_Products_Sold'})

mostproducts_sold=atheletic_sales_pivot.sort_values('Total_Products_Sold',ascending=False)
# Show the top 5 results.
mostproducts_sold.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

athletic_sales_new = athletic_sales_2021_2021_womensfootwear.rename(columns={'total_sales':'Total Sales'})
atheletic_pivot = pd.pivot_table(athletic_sales_new,values=['Total Sales'],index=['invoice_date'],aggfunc='sum')

atheletic_pivot=atheletic_pivot.sort_values(by=['Total Sales'],ascending=False)
atheletic_pivot.head(10)

# Optional: Rename the "total_sales" column to "Total Sales"
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# Sort the resampled pivot table in ascending order on "Total Sales".

# Define the year
start_date = pd.Timestamp(2020, 1, 1)
end_date = pd.Timestamp(2021, 12, 31)
date_range = pd.date_range(start=start_date, end=end_date)

atheletic_pivot['date_bin'] = pd.cut(atheletic_pivot.index, bins=date_range)

total_sales_per_bin = atheletic_pivot.groupby('date_bin')['Total Sales'].sum()

total_sales_per_bin=total_sales_per_bin.sort_values(ascending=False)

total_sales_per_bin.head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.


# Sort the resampled pivot table in ascending order on "Total Sales".
```

```python
# Define the year
start_date = pd.Timestamp(2020, 1, 1)
end_date = pd.Timestamp(2021, 12, 31)
date_range = pd.date_range(start=start_date, end=end_date,freq='W')

atheletic_pivot['date_bin'] = pd.cut(atheletic_pivot.index, bins=date_range)

total_sales_per_bin = atheletic_pivot.groupby('date_bin')['Total Sales'].sum()

total_sales_per_bin=total_sales_per_bin.sort_values(ascending=False)

total_sales_per_bin.head(10)
```
